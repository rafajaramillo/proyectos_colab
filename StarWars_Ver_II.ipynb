```python
import pandas as pd
import requests
import json
```

```python
def GetStarWarsCharacters(link):
  datos = requests.get(link)
  api_results = json.loads(datos.content)
  for character in api_results['results']:
    #if 'https://swapi.dev/api/people/6/' in character ['people']:
      #yield character['name']
      yield character
  if 'next' in api_results and api_results['next'] is not None:
    next_page = GetStarWarsCharacters(api_results['next'])
    for page in next_page:
      yield page
```

```python
persons = GetStarWarsCharacters('https://swapi.dev/api/people')
character_df1 = pd.DataFrame(persons)
#for result in persons:
  #print(result)
```
