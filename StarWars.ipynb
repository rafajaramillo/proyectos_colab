```python
import pandas as pd
import requests
```

```python
# function to get star wars characters
def GetStarWarsCharacters():
    response = requests.get('https://swapi.dev/api/people/')
    #response = requests.get('https://swapi.py4e.com/api/people/')
    data_dict = response.json()
    characters = data_dict['results']
    return characters
```

```python
def GetStarWarsCharacters(pagina):
    parametros = {'page': pagina}  
    response = requests.get('https://swapi.dev/api/people/', params=parametros)
    #response = requests.get('https://swapi.py4e.com/api/people', params=parametros)
    data_dict = response.json()
    characters = data_dict['results']
    return characters
```

```python
for n in range(9):
  characters = GetStarWarsCharacters(n+1) # This function gets the people as json
  character_df1 = pd.DataFrame(characters)  # convertimos el json en un DataFrame
  if n == 1:
    df_concat_Starwars = character_df1
  if n > 1:
    df_concat_Starwars = pd.concat([df_concat_Starwars, character_df1]) # aquí se concatena cada DataFrame
  
```

```python
# Vamos a filtrar sólo las columnas necesarias del DataFrame
filterSW_df = df_concat_Starwars[['name','height','mass','hair_color','skin_color', 'eye_color','birth_year','gender']]
filterSW_df
```
